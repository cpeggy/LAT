```python
resultS = Q1_ansm.loc[Q1_ansm['地區'] == '南部']
print(resultS)
print("南部有",len(resultS),"所學校重複成功申請")
```

```python
resultE = Q1_ansm.loc[Q1_ansm['地區'] == '東部']
print(resultE)
print("東部有",len(resultE),"所學校重複成功申請")
```

```python
#比例
NN = len(resultN) / len(allapplyschool) * 100
CC = len(resultC) / len(allapplyschool) * 100
SS = len(resultS) / len(allapplyschool) * 100
EE = len(resultE) / len(allapplyschool) * 100

print("北部: {:.2f}%".format(NN))
print("中部: {:.2f}%".format(CC))
print("南部: {:.2f}%".format(SS))
print("東部: {:.2f}%".format(EE))
```
